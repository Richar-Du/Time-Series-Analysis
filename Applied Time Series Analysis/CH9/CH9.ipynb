### Example 9.1

回归模型（9.1）适用于平方根变换的英格兰和威尔士每月降雨量序列，其中$\epsilon_t$被认为是白噪声，该假设被诊断测试支持。 由于已经进行了降雨的平方根转换，因此第i个月的平均降雨量由$\alpha_i^2$给出，其估计值为如书中给出。季节性模式十分明显： 最干旱的季节是2月至6月，平均每月602毫米，而7月至1月的平均时间则增加了近40％，为834毫米。

```python
# 导入第三方包
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
from math import log,sqrt
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
rainfall=pd.read_csv("rainfall.csv")['rainfall']

def season_mon(month,time):
    if(time%12==month):
        return 1
    else:
        return 0


def estimate_rainfall(sample,p):  
    matrix_x=np.zeros((len(sample)-p,p+12))  
    array=sample.reshape(len(sample))  #array是样本全体
    j=0
    for i in range(0,len(sample)-p):
        matrix_x[i,0:p+12]=np.hstack((np.array([season_mon(1,p+i+1),season_mon(2,p+i+1),season_mon(3,p+i+1),season_mon(4,p+i+1),season_mon(5,p+i+1),season_mon(6,p+i+1),season_mon(7,p+i+1),season_mon(8,p+i+1),season_mon(9,p+i+1),season_mon(10,p+i+1),season_mon(11,p+i+1),season_mon(0,p+i+1)]),array[j:j+p]))
        j+=1  
    matrix_y=np.array(array[p:len(sample)])
    matrix_y=matrix_y.reshape(len(sample)-p,1)
    #print(matrix_x)
    #所有系数的表达式：A=(X^TX)^-1 X^T Y
    coef=np.dot(np.dot(np.linalg.inv(np.dot(matrix_x.T,matrix_x)),matrix_x.T),matrix_y)
    return coef

para=estimate_rainfall(np.array(rainfall),0)
#print(para)
alpha2=[]
for i in range(0,12):
    alpha2.append(int(para[i][0]**2*10))
print(alpha2)
```

```python
print('二到六月平均降雨量：',sum(alpha2[1:6])/5)
```

```python
print('七到一月平均降雨量：',(sum(alpha2[6:12])+alpha2[0])/7)
```

由于没有证据表明这个简单的模型被误拟合，因此进一步得出的结论是：这种季节性模式在整个140多年的采样期内一直保持不变。

### Example 9.2

啤酒销售量的$\triangledown \triangledown^4$变换的ACF已经给出，可以计算得到$r_1...r_5$如下所示：

```python
# 读入数据
beer=pd.read_csv("beer.csv")['beer']
```

```python
#### 公共函数：差分
import statsmodels.tsa.api as smt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
def delta_serie(vector):
    ret=[]
    for i in range(1,len(vector)):
        ret.append(vector[i]-vector[i-1])
    return ret

def acf(ts, k):
    x = np.array(ts) - np.mean(ts)
    coeff = np.zeros(k+1, np.float64) # to store acf
    coeff[0] = x.dot(x) # N*c(0)
    for i in range(1, k+1):
        coeff[i] = x[:-i].dot(x[i:]) # (N-k)*c(i)        
    return coeff / coeff[0]-0.1

#生成wt
def delta_beer(vector):
    ret=[]
    for i in range(5,len(vector)):
        ret.append(vector[i]-vector[i-1]-vector[i-4]+vector[i-5])
    return ret


#d1_beer=delta_quarter(beer)
beer_final=delta_beer(beer)


acf_30=acf(beer_final,30).tolist()
#print('acf值:',acf_30)
#画出k=1,2,...,30的acf
plt.plot(acf_30,label='acf')


vectornp=np.array(beer_final)
plot_acf(vectornp).show()
print('r1=%f,r2=%f,r3=%f\nr4=%f,r5=%f'%(round(acf_30[1],2),round(acf_30[2],2),round(acf_30[3],2),round(acf_30[4],2),round(acf_30[5],2)))
```

由于$r_2$近似为0，$r_1  r_4=0.36$，前五个样本的自相关函数是在合理的范围之内。

### Example 9.3

在Example 4.3中，将无漂移的ARIMA（0,1,3）过程拟合到每月全球温度，在Example7.2中，使用该模型提供了到2020年的预测。由于$\theta_2$和$\theta_3$虽然很显着，但在 与$\theta_1$相比，ARIMA（0,1,1）过程应能很好地拟合，实际上，$\theta= 0.55$且均方根误差（RMSE）为0.1255 [ARIMA（ 0，1，3）]。 从简单指数平滑和ARIMA（0，1，1）的等价性来看，我们希望前者模型产生相似的拟合并预测平滑参数$\alpha= 0.45$。 通过简单的指数平滑和估计$\alpha$拟合序列确实会导致平滑参数的该值，RMSE为0.1257，并且预测由$f_{T,h} = z_T = 0.581$给出。 这些应与Example 7.2中获得的ARIMA（0,1,3）预测进行比较，对于h> 2，它们等于0.621。

```python
temp=pd.read_csv("Global temperature.csv")['value']
```

```python
def RMSE(vector1,vector2):
    s=0
    for i in range(len(vector1)):
        s+=(vector1[i]-vector2[i])**2
    return s

# Simple Exponential Smoothing
fit1 = SimpleExpSmoothing(temp).fit(smoothing_level=0.2,optimized=False)
# plot
#l1, = plt.plot(list(fit1.fittedvalues) + list(fit1.forecast(5)))
fit2 = SimpleExpSmoothing(temp).fit(smoothing_level=0.3,optimized=False)
# plot
#l2, = plt.plot(list(fit2.fittedvalues) + list(fit2.forecast(5)))
fit3 = SimpleExpSmoothing(temp).fit(smoothing_level=0.4,optimized=False)
# plot
#l3, = plt.plot(list(fit3.fittedvalues) + list(fit3.forecast(5)))
fit4 = SimpleExpSmoothing(temp).fit(smoothing_level=0.45,optimized=False)
fit5 = SimpleExpSmoothing(temp).fit(smoothing_level=0.5,optimized=False)
#l4, = plt.plot(temp)
#plt.legend(handles = [l1, l2, l3, l4], labels = ['a=0.2', 'a=0.6', 'auto', 'data'], loc = 'best', prop={'size': 7})
#plt.show()
v1=list(fit1.fittedvalues)
v2=list(fit2.fittedvalues)
v3=list(fit3.fittedvalues)
R=[RMSE(v1,temp),RMSE(v2,temp),RMSE(v3,temp),RMSE(v4,temp),RMSE(v5,temp),RMSE(v6,temp),RMSE(v7,temp),RMSE(v8,temp),RMSE(v9,temp)]
x=[0.2,0.3,0.4,0.45,0.5,0.6,0.7,0.8,0.9]
plt.plot(x,R,marker='.')
plt.title('RMSE with alternative alpha')
plt.xlabel('alpha')
plt.ylabel('RMSE')
plt.show()
fit41 = SimpleExpSmoothing(temp).fit(smoothing_level=0.41,optimized=False)
fit42 = SimpleExpSmoothing(temp).fit(smoothing_level=0.42,optimized=False)
fit43 = SimpleExpSmoothing(temp).fit(smoothing_level=0.43,optimized=False)
v41=list(fit41.fittedvalues)
v42=list(fit42.fittedvalues)
v43=list(fit43.fittedvalues)
R1=[RMSE(v41,temp),RMSE(v42,temp),RMSE(v43,temp),RMSE(v44,temp),RMSE(v45,temp),RMSE(v46,temp),RMSE(v47,temp),RMSE(v48,temp),RMSE(v49,temp)]
x1=[0.41,0.42,0.43,0.44,0.45,0.46,0.47,0.48,0.49]
plt.plot(x1,R1,marker='.')
plt.title('RMSE when alpha in (0.4,0.5)')
plt.xlabel('alpha')
plt.ylabel('RMSE')
plt.show()
print('最优alpha处的预测值:',fit4.forecast(1))
```

要认识到全球温度呈线性趋势的可能性，将需要使用双指数平滑法或Holt-Winters。 前者估计单个平滑参数为$\gamma=0.196$，RMSE=0.1319。因此，Holt-Winters预测包括一个小的正线性趋势，到2020年12月预测期结束时，预测值将从0.582增加到0.599。在这两种情况下，都没有明显的正漂移出现，与我们先前的结果一致。 鉴于Holt-Winters的均方根误差（RMSE）为0.1256，这意味着简单的指数平滑法是这三种预测每月全球温度的技术中最合适的。
