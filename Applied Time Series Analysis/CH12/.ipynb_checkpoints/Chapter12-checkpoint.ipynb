### Example 12.1

```python
# 导入需要的包
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels
from statsmodels.tsa.ar_model import AutoReg
import statsmodels.tsa.api as smt
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
r20=pd.read_csv("r20.csv").dropna(axis=0,how='any')['r20']
rs=pd.read_csv("rs.csv").dropna(axis=0,how='any')['rs']
plt.plot(r20,label='r20')
plt.plot(rs,'r',linestyle='--',label='rs')
plt.legend()
plt.title('r20 and rs')
plt.show()
```

在Example3.2中，我们根据ACF和PACF，可以确定R20一阶差分后的序列为AR(2)过程。

```python
def ar_estimation(sample,p):  
    matrix_x=np.zeros((len(sample)-p,p))  
    array=sample.reshape(len(sample))  
    j=0
    for i in range(0,len(sample)-p):
        matrix_x[i,0:p]=array[j:j+p]  
        j+=1  
    matrix_y=np.array(array[p:len(sample)])
    matrix_y=matrix_y.reshape(len(sample)-p,1)
    #AR系数的表达式：A=(X^TX)^-1 X^T Y
    coef=np.dot(np.dot(np.linalg.inv(np.dot(matrix_x.T,matrix_x)),matrix_x.T),matrix_y)
    return coef
def delta_serie(vector):
    ret=[]
    for i in range(1,len(vector)):
        ret.append(vector[i]-vector[i-1])
    return ret
dr20=delta_serie(r20)
co=ar_estimation(np.array(dr20),2)
print(co)
print('均值为:',np.mean(dr20))
print('白噪声标准差为:',np.std(dr20)/((1+co[0][0]**2)*(1+co[1][0]**2)))
std2=np.std(r20)
var=std2**2
```

分别计算不同阶数的ARDL模型的AIC统计量，从而确定阶数。

```python
#用最小二乘法估计ARDL(p,q)模型：
def ARDLpq_estimation(sample,samplex,p,q):
    if len(sample)!=len(samplex):
        print('Wrong!')
        return
    h=max(p,q)
    matrix_x=np.zeros((len(sample)-h,p+q+1))  
    array=sample.reshape(len(sample))
    arrayx=samplex.reshape(len(sample))
    j=0
    for i in range(0,len(sample)-h):
        matrix_x[i,0:p+q+1]=np.hstack((array[j:j+p],arrayx[j:j+q+1]))
        j+=1  
    matrix_y=np.array(array[h:len(sample)])
    matrix_y=matrix_y.reshape(len(sample)-h,1)
    #系数的表达式：A=(X^TX)^-1 X^T Y
    coef=np.dot(np.dot(np.linalg.inv(np.dot(matrix_x.T,matrix_x)),matrix_x.T),matrix_y)
    return coef
drs=delta_serie(rs)
parapq=ARDLpq_estimation(np.array(dr20),np.array(drs),2,2)
print(ARDLpq_estimation(np.array(dr20),np.array(drs),2,2))
print('白噪声标准差估计为：',np.std(dr20)/((1+parapq[0][0]**2)*(1+parapq[1][0]**2)*(1+parapq[2][0]**2)*(1+parapq[3][0]**2)*(1+parapq[4][0]**2)))
```

根据最终结果可以看出，ARDL(2,2)模型是：
$DR20_t=0.230DR20_{t-1}-0.120DR20_{t-2}+0.296DRS_t-0.076DRS_{t-1}+0.016DRS_{t-2},\space \space \sigma_a=0.249$

### Example 12.2

图12.2显示了全球温度和四个可能的诱导因素，即人为和自然因素造成的总辐射力（不包括火山喷发），平流层气溶胶引起的辐射力以及南方涛动和大西洋多年代际涛动的比例指数。 从这个数字可以看出，全球温度和总辐射力显然是不稳定的，单位根测试证实它们都是I（1）过程。 然而，火山力和两种海洋-大气现象显然是静止的，这也得到了单位根试验的证实。

所有的外界因素都看作外源变量，令m=6，我们选择ARDL(3,2,0,1,4)模型，下面对模型的参数进行估计。

```python
# 读入数据
amo=np.read_csv("amo.csv")['amo']
soi=np.read_csv("soi.csv")['soi']
temp=np.read_csv("temp.csv")['temp']
trf=np.read_csv("trf.csv")['trf']
volc=np.read_csv("volc.csv")['volc']
dtemp=delta_serie(temp)
dtrf=delta_serie(trf)
```

```python
def ARDLpqabc_estimation(sample,samplex1,samplex2,samplex3,samplex4,p,q,a,b,c):
    h=max(p,q,a,b,c)
    matrix_x=np.zeros((len(sample)-h,p+q+a+b+c+5))  
    array=sample.reshape(len(sample))
    arrayx1=samplex1.reshape(len(sample))
    arrayx2=samplex2.reshape(len(samplex2))
    arrayx3=samplex3.reshape(len(samplex3))
    arrayx4=samplex4.reshape(len(samplex4))
    j=0
    for i in range(0,len(sample)-h):
        matrix_x[i,0:p+q+a+b+c+5]=np.hstack(([1],array[j:j+p],arrayx1[j:j+q+1],arrayx2[j:j+a+1],arrayx3[j:j+b+1],arrayx4[j:j+c+1]))
        j+=1  
    matrix_y=np.array(array[h:len(sample)])
    matrix_y=matrix_y.reshape(len(sample)-h,1)
    #系数的表达式：A=(X^TX)^-1 X^T Y
    coef=np.dot(np.dot(np.linalg.inv(np.dot(matrix_x.T,matrix_x)),matrix_x.T),matrix_y)
    return coef


def estimate_DF(sample,p):  
    matrix_x=np.zeros((len(sample)-p,p+1))  
    array=sample.reshape(len(sample))  #array是样本全体
    j=0
    for i in range(0,len(sample)-p):
        matrix_x[i,0:p+1]=np.hstack((np.array([1]),array[j:j+p]))
        j+=1  
    #在原有AR的OLS方法基础上加上beta0+beta1*t这一项
    matrix_y=np.array(array[p:len(sample)])
    matrix_y=matrix_y.reshape(len(sample)-p,1)
    #所有系数的表达式：A=(X^TX)^-1 X^T Y
    coef=np.dot(np.dot(np.linalg.inv(np.dot(matrix_x.T,matrix_x)),matrix_x.T),matrix_y)
    return coef



paras=ARDLpqabc_estimation(np.array(dtemp),np.array(dtrf),np.array(volc),np.array(soi),np.array(amo),3,2,0,1,4)
print(ARDLpqabc_estimation(np.array(dtemp),np.array(dtrf),np.array(volc),np.array(soi),np.array(amo),3,2,0,1,4))
p=[]
for v in paras:
    p.append(v[0])
sum=1
for item in p:
    sum*=(1+item**2)
print('白噪声估计值：',np.std(dtemp)/sum)
```

上面的列表分别表示公式12.9中的参数：

$\beta_0,\phi_3,\phi_2,\phi_1,\alpha_2,\alpha_1,\alpha_0,\beta_0,\gamma_1,\gamma_0,\rho_4,\rho_3,\rho_2,\rho_1,\rho_0$

除了$\beta_4$，所有的参数都显著。

从上述模型可以看出，总辐射强度的水平对气温有一个正向的影响，而火山爆发和SO对气温的影响是逐渐累积的。同时，也可得知AMO对气温的长期影响是0。
